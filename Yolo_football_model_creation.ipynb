```python
import os
import shutil
import random
from tqdm.notebook import tqdm
```

```python
pwd
```

```python
train_path_img="./yolo_data/images/train/"
train_path_label="./yolo_data/labels/train/"
val_path_img="./yolo_data/images/val/"
val_path_label="./yolo_data/labels/val/"
```

```python
random.seed(42)
```

```python
def train_test_split(path,split=0.2) :

  files=list(set([i[:-4] for i in os.listdir(path)]))
  # print (files)
  # print (len(files))
  random.shuffle(files)
  test_size=int(len(files)*split)
  train_size=len(files)-test_size

  os.makedirs(train_path_img,exist_ok=True)
  os.makedirs(train_path_label,exist_ok=True)
  os.makedirs(val_path_img,exist_ok=True)
  os.makedirs(val_path_label,exist_ok=True)

  for file in tqdm(files[:train_size]) :

    shutil.copy2(path+"/"+file+".jpg",train_path_img+file+".jpg")
    shutil.copy2(path+"/"+file+".txt",train_path_label+file+"txt")
  # print (len(os.listdir(train_path_img)))
  # print (len(os.listdir(train_path_label)))

  for file in tqdm(files[train_size:]) :

    shutil.copy2(path+"/"+file+".jpg",val_path_img+file+".jpg")
    shutil.copy2(path+"/"+file+".txt",val_path_label+file+".txt")

train_test_split("/content/drive/MyDrive/yolo/data")
```
