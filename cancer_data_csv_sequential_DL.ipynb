```python
import pandas as pd

X=pd.read_csv("/content/drive/MyDrive/data sets/cancer_data.csv")
X
```

```python
print (X.corr())
```

```python
print (X.isna().sum())
```

```python
print (X.dtypes)
```

```python
x=X.iloc[:,0:5]
x
```

```python
y=X.iloc[:,5]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
z=MinMaxScaler()
z.fit(x)
w=z.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(w,y,test_size=0.3,random_state=1)
```

```python
from keras.models import Sequential
q=Sequential()
```

```python
from keras.layers import Dense

q.add(Dense(5,activation="relu"))
q.add(Dense(7,activation="relu"))
q.add(Dense(9,activation="relu"))
q.add(Dense(1,activation="sigmoid"))
```

```python
q.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
q.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python
q.summary()
```

```python
q.evaluate(x_test,y_test)
```

```python
l=q.predict(z.transform([[17.99,10.38,122.80,1001.0,0.11840]]))

if l >= 0.5 :

  print ("1")

else :

  print ("0")
```
