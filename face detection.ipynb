Image Face Detection

```python
import cv2 
import matplotlib.pyplot as plt 

img=cv2.imread(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\face detection\fam.webp") 
plt.imshow(img)
```

```python
fdm=cv2.CascadeClassifier(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\face detection\haarcascade_frontalface_default.xml") 
fdm
```

```python
gray=cv2.imread(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\face detection\fam.webp",cv2.IMREAD_GRAYSCALE) 
gray
```

```python
gray2=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) 
gray2
```

```python
face_detection=fdm.detectMultiScale(gray) 
face_detection 
```

```python
for x,y,h,w in face_detection : 

    cv2.rectangle(img2,pt1=(x,y),pt2=(x+w,y+h),color=(100,150,200),thickness=4) 

plt.imshow(img2) 
```

Video Face Detection

```python
import cv2
video=cv2.VideoCapture(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\face detection\video.mp4") 

while True : 

    sucess,frame=video.read() 
    if not sucess : 

        break 
    fdm=cv2.CascadeClassifier(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\face detection\haarcascade_frontalface_default.xml") 
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 
    face_detection=fdm.detectMultiScale(gray) 
    for x,y,h,w in face_detection : 

        cv2.rectangle(frame,(x,y),(x+w,y+h),color=(255,255,255),thickness=3) 

    cv2.imshow("face detection",frame) 
    if cv2.waitKey(1) & 0XFF==ord("q") : 

        break 

video.release() 
cv2.destroyAllWindows() 
```

Web cam Face Detection

```python
import cv2 

webcam=cv2.VideoCapture(0) 

while True : 
    
    sucess,frame=webcam.read() 
    fdm=cv2.CascadeClassifier(r"C:\Users\shibu\OneDrive\Documents\Deep learning workout\work out\haarcascade_frontalface_default.xml") 
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 
    faces=fdm.detectMultiScale(gray) 
    
    for x,y,w,h in faces : 
        
        cv2.rectangle(frame,(x,y),(x+w,y+h),color=(255,255,255),thickness=4) 
        
    cv2.imshow("web cam",frame) 
    
    if cv2.waitKey(1) & 0XFF==ord("q") : 
        
        break 
    
webcam.release() 
cv2.destroyAllWindows() 
```
